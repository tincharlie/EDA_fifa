# Data Cleaning

<h3 style="color:brown" align="center">Filling the missing value for the continous variables for proper data visualization
</h3>


```python
A['ShortPassing'].fillna(A['ShortPassing'].mean(), inplace = True)
A['Volleys'].fillna(A['Volleys'].mean(), inplace = True)
A['Dribbling'].fillna(A['Dribbling'].mean(), inplace = True)
A['Curve'].fillna(A['Curve'].mean(), inplace = True)
A['FKAccuracy'].fillna(A['FKAccuracy'], inplace = True)
A['LongPassing'].fillna(A['LongPassing'].mean(), inplace = True)
A['BallControl'].fillna(A['BallControl'].mean(), inplace = True)
A['HeadingAccuracy'].fillna(A['HeadingAccuracy'].mean(), inplace = True)
A['Finishing'].fillna(A['Finishing'].mean(), inplace = True)
A['Crossing'].fillna(A['Crossing'].mean(), inplace = True)
A['Weight'].fillna('200lbs', inplace = True)
A['Contract Valid Until'].fillna(2019, inplace = True)
A['Height'].fillna("5'11", inplace = True)
A['Loaned From'].fillna('None', inplace = True)
A['Joined'].fillna('Jul 1, 2018', inplace = True)
A['Jersey Number'].fillna(8, inplace = True)
A['Body Type'].fillna('Normal', inplace = True)
A['Position'].fillna('ST', inplace = True)
A['Club'].fillna('No Club', inplace = True)
A['Work Rate'].fillna('Medium/ Medium', inplace = True)
A['Skill Moves'].fillna(A['Skill Moves'].median(), inplace = True)
A['Weak Foot'].fillna(3, inplace = True)
A['Preferred Foot'].fillna('Right', inplace = True)
A['International Reputation'].fillna(1, inplace = True)
A['Wage'].fillna('€200K', inplace = True)
```

```python
A.fillna(0, inplace = True)
```

#### Taking Mean Of different quality with uniqueness columns

```python
def defending(df):
    return int(round((df[["Marking", "StandingTackle", 'SlidingTackle']].mean()).mean()))
def general(df):
    return int(round((df[["HeadingAccuracy","Dribbling", "Curve", 'BallControl']].mean()).mean()))
def mental(df):
    return int(round((df[["Interceptions","Aggression", "Positioning", 'Composure', 'Vision']].mean()).mean()))
def passing(df):
    return int(round((df[["LongPassing", "ShortPassing", 'Crossing']].mean()).mean()))
def mobility(df):
    return int(round((df[["Acceleration", "SprintSpeed", "Agility", "Reactions"]].mean()).mean()))
def power(df):
    return int(round((df[["Balance", "Jumping", "Stamina", "Strength"]].mean()).mean()))
def rating(df):    
    return int(round((df[["Potential", "Overall"]].mean()).mean()))
def shooting(df):
    
    return int(round((df[["Finishing","Volleys","FKAccuracy", "ShotPower", "LongShots", "Penalties"]].mean()).mean()))
```

<h4 style="color:brown" align="center">Now We are defining a function for cleaning the Weight data </h4>


```python
def extract_value_from(value):
  out = value.replace('lbs', '')
  return float(out)


A['Weight'] = A['Weight'].apply(lambda x : extract_value_from(x))

A['Weight'].head()
```

<h3 style="color:brown" align="center">defining a function for cleaning the wage column</h3>


```python
def extract_value_from(Value):
    out = Value.replace('€', '')
    if 'M' in out:
        out = float(out.replace('M', ''))*1000000
    elif 'K' in Value:
        out = float(out.replace('K', ''))*1000
    return float(out)
```

<h3 style="color:brown" align="center">We are applying the function to the wage column</h3>


```python
A['Value'] = A['Value'].apply(lambda x: extract_value_from(x))
A['Wage'] = A['Wage'].apply(lambda x: extract_value_from(x))

A['Wage'].head()
```
